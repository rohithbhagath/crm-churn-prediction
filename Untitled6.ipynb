```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import warnings
warnings.filterwarnings('ignore')

from sklearn.metrics import (
    accuracy_score, precision_score,
    recall_score, f1_score,
    roc_auc_score, confusion_matrix,
    classification_report, roc_curve,
    precision_recall_curve, average_precision_score
)
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import cross_val_score, StratifiedKFold

# Load dataset
df = pd.read_csv('/content/engineered_churn_data.csv')

# Load best model & scaler
best_model = joblib.load('/content/best_churn_model.pkl')
scaler     = joblib.load('/content/scaler.pkl')

print("✅ Model & Scaler Loaded!")
print("Model:", best_model)
```

```python
# Encode categorical columns
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
le = LabelEncoder()

for col in cat_cols:
    df[col] = le.fit_transform(df[col].astype(str))

# Features & Target
X = df.drop(columns=['Churn'])
y = df['Churn']

# Scale
X_scaled = scaler.transform(X)

# Predictions
y_pred  = best_model.predict(X_scaled)
y_proba = best_model.predict_proba(X_scaled)[:, 1]

print("✅ Data Prepared!")
print("Total Samples:", len(y))
```

```python
# Calculate all metrics
accuracy  = accuracy_score(y, y_pred)
precision = precision_score(y, y_pred)
recall    = recall_score(y, y_pred)
f1        = f1_score(y, y_pred)
roc_auc   = roc_auc_score(y, y_proba)

print("=" * 45)
print("       MODEL EVALUATION REPORT")
print("=" * 45)
print(f"  Accuracy  : {accuracy:.4f}  ({accuracy*100:.2f}%)")
print(f"  Precision : {precision:.4f}  ({precision*100:.2f}%)")
print(f"  Recall    : {recall:.4f}  ({recall*100:.2f}%)")
print(f"  F1 Score  : {f1:.4f}  ({f1*100:.2f}%)")
print(f"  ROC-AUC   : {roc_auc:.4f}  ({roc_auc*100:.2f}%)")
print("=" * 45)
```

```python
print("\n📋 Detailed Classification Report:")
print("=" * 45)
print(classification_report(
    y, y_pred,
    target_names=['No Churn', 'Churn']
))
```

```python
cm = confusion_matrix(y, y_pred)

plt.figure(figsize=(7, 5))
sns.heatmap(cm,
            annot=True,
            fmt='d',
            cmap='Blues',
            xticklabels=['No Churn', 'Churn'],
            yticklabels=['No Churn', 'Churn'])

plt.title('Confusion Matrix', fontsize=14)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.tight_layout()
plt.show()

# Extract values
tn, fp, fn, tp = cm.ravel()
print(f"\nTrue Positives  (Churn correctly identified) : {tp}")
print(f"True Negatives  (Retained correctly)         : {tn}")
print(f"False Positives (Wrongly flagged as Churn)   : {fp}")
print(f"False Negatives (Missed Churners)            : {fn}")
```

```python
fpr, tpr, thresholds = roc_curve(y, y_proba)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange',
         lw=2, label=f'ROC Curve (AUC = {roc_auc:.3f})')
plt.plot([0, 1], [0, 1], color='navy',
         lw=2, linestyle='--', label='Random Classifier')
plt.fill_between(fpr, tpr, alpha=0.1, color='darkorange')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-AUC Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
precision_vals, recall_vals, _ = precision_recall_curve(y, y_proba)
avg_precision = average_precision_score(y, y_proba)

plt.figure(figsize=(8, 6))
plt.plot(recall_vals, precision_vals,
         color='green', lw=2,
         label=f'PR Curve (AP = {avg_precision:.3f})')
plt.fill_between(recall_vals, precision_vals,
                 alpha=0.1, color='green')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='upper right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 5-Fold Stratified Cross Validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_scores = cross_val_score(
    best_model, X_scaled, y,
    cv=cv, scoring='f1'
)

print("\n📊 Cross Validation Results (5-Fold):")
print("=" * 45)
for i, score in enumerate(cv_scores):
    print(f"  Fold {i+1}: F1 Score = {score:.4f}")
print(f"\n  Mean F1  : {cv_scores.mean():.4f}")
print(f"  Std Dev  : {cv_scores.std():.4f}")
print("=" * 45)

# Visualize
plt.figure(figsize=(8, 5))
plt.bar(range(1, 6), cv_scores,
        color='steelblue', edgecolor='black')
plt.axhline(cv_scores.mean(), color='red',
            linestyle='--', label=f'Mean F1: {cv_scores.mean():.4f}')
plt.xlabel('Fold')
plt.ylabel('F1 Score')
plt.title('5-Fold Cross Validation F1 Scores')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Find optimal threshold
thresholds_range = np.arange(0.1, 0.9, 0.05)
f1_scores = []

for thresh in thresholds_range:
    y_pred_thresh = (y_proba >= thresh).astype(int)
    f1_scores.append(f1_score(y, y_pred_thresh))

optimal_threshold = thresholds_range[np.argmax(f1_scores)]
optimal_f1        = max(f1_scores)

plt.figure(figsize=(10, 5))
plt.plot(thresholds_range, f1_scores,
         color='purple', lw=2, marker='o')
plt.axvline(optimal_threshold, color='red',
            linestyle='--',
            label=f'Optimal Threshold: {optimal_threshold:.2f}')
plt.xlabel('Threshold')
plt.ylabel('F1 Score')
plt.title('F1 Score vs Decision Threshold')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(f"\n✅ Optimal Threshold : {optimal_threshold:.2f}")
print(f"✅ Best F1 Score     : {optimal_f1:.4f}")
```

```python
plt.figure(figsize=(10, 5))

sns.histplot(
    y_proba[y == 0],
    bins=50, color='green',
    label='No Churn', alpha=0.6, kde=True
)
sns.histplot(
    y_proba[y == 1],
    bins=50, color='red',
    label='Churn', alpha=0.6, kde=True
)

plt.axvline(optimal_threshold, color='black',
            linestyle='--',
            label=f'Threshold: {optimal_threshold:.2f}')
plt.xlabel('Churn Probability')
plt.ylabel('Count')
plt.title('Churn Probability Distribution')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
print("\n")
print("=" * 55)
print("          FINAL MODEL EVALUATION SUMMARY")
print("=" * 55)
print(f"  Model              : {type(best_model).__name__}")
print(f"  Total Customers    : {len(y)}")
print(f"  Churned Customers  : {y.sum()} ({y.mean()*100:.2f}%)")
print("-" * 55)
print(f"  Accuracy           : {accuracy*100:.2f}%")
print(f"  Precision          : {precision*100:.2f}%")
print(f"  Recall             : {recall*100:.2f}%")
print(f"  F1 Score           : {f1*100:.2f}%")
print(f"  ROC-AUC            : {roc_auc*100:.2f}%")
print(f"  CV Mean F1         : {cv_scores.mean()*100:.2f}%")
print("-" * 55)
print(f"  Optimal Threshold  : {optimal_threshold:.2f}")
print(f"  True Positives     : {tp}")
print(f"  False Negatives    : {fn} (Missed Churners)")
print(f"  False Positives    : {fp} (Wrong Alerts)")
print("=" * 55)
```

```python
# Save summary as CSV
report_df = pd.DataFrame({
    'Metric': ['Accuracy', 'Precision', 'Recall',
                'F1 Score', 'ROC-AUC', 'CV Mean F1'],
    'Score' : [accuracy, precision, recall,
                f1, roc_auc, cv_scores.mean()]
}).round(4)

report_df.to_csv('model_evaluation_report.csv', index=False)
print("✅ Evaluation report saved: model_evaluation_report.csv")
```
